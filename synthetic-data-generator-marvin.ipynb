```python
#GENERATE CREDIT CARD TRANSACTIONS - MILLIONS OF RECORDS
#--------------------------
from faker import Faker
import pandas as pd
fake = Faker(locale='en_PH')

def create_cc_transactions(data):
    #create an empty list to add our cc_transaction dictionaries
    data_list = []
    for i in range (1, data):
        cc_transaction = {}
        cc_transaction['ACCOUNT_NUMBER'] = fake.bban()
        cc_transaction['AMOUNT'] = fake.random_int (min=1000, max=150000, step=500)
        cc_transaction['CHANNEL'] = fake.random_element(elements=  ['POS', 'Online', "ATM", "Mobile App", "Branch"])
        cc_transaction['CURRENCY'] = fake.random_element(elements=  ['PHP', 'SGD', 'USD','AUD'])
        cc_transaction['MERCHANT_CATEGORY_CODE'] = fake.random_number(4)
        cc_transaction['MERCHANT_NAME'] = fake.company()
        cc_transaction['TRANSACTION_DATETIME'] = fake.date_time_this_year()
        cc_transaction['TRANSACTION_TYPE'] = fake.random_element(elements=  ['Purchase', 'Withdrawal', "Payment", "Refund"])
        data_list.append (cc_transaction)

    return pd.DataFrame(data_list)

data = create_cc_transactions(100)
data.to_csv("credit card transactions.csv")
```
